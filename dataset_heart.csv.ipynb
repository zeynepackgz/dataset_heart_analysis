```python
age_filter= df[df["age"]<50] #50 yaşından küçük hastaları göster
age_filter
```

```python
age_filter["sex"]
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x="sex",y="age",color='pink',data=age_filter)
plt.show()
```
